```python
def plotter(ax,data1,data2,dict1):
    out=ax.plot(data1,data2,**dict1)
    return out
data1,data2,data3,data4=np.random.randn(4,100)
fig,ax=mt.subplots(1,1)
plotter(ax,data1,data2,{"marker":"x"})
mt.show()
```

```python
fig,(ax1,ax2)=mt.subplots(1,2)
plotter(ax1,data3,data4,{'marker':'x'})
plotter(ax2,data1,data2,{'marker':'o'})
mt.show()
```
