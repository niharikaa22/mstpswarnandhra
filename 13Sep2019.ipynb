#### Linear graph

```python
import matplotlib.pyplot as plt

plt.title("Basic Linear Graph")
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph a plot between distance and time

distance=list(range(2,11,2))
time=list(range(1,6))
#d=2*t
plt.title("Distance vs Time")
plt.xlabel("time")
plt.ylabel("distance")
plt.plot(time,distance,'go')
plt.grid()
plt.show()
```

```python
# v=a t
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
# plot a graph for number vs squares

n=list(range(30))
s=list(i**2 for i in n)
plt.plot(n,s)
plt.grid()
plt.show()
```

```python
# voltage vs time sinusoidal graph
import numpy as np
t=np.arange(0.1,2.0,0.1)
v=[1+np.sin(2*np.pi*i) for i in t]
plt.plot(t,v)
plt.grid()
plt.xlabel('time')
plt.ylabel("voltage")
plt.title("Time Vs Voltage Graph")
plt.savefig("voltage-time.png")
plt.show()
```
